__This graphs represent the Major changed or who had first major in It or Computer science feild__

```python
value_counts = major['First Major Choice'].value_counts().reset_index()
value_counts.columns = ['Major', 'Count']
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=value_counts, x='Major', y='Count')  # Removed palette
plt.title('Count of Majors')
plt.xlabel('Major')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x labels if needed
plt.show()
```

```python
value_counts = major['Financial Aid'].value_counts()
value_counts.plot(kind='bar', color='skyblue')
plt.title('Financial Aid Distribution')
plt.xlabel('Financial Aid Status')
plt.ylabel('Count')  
plt.show()
```

```python
value_counts = major['What motivated you to seek a computing degree/certificate at CCM?  [To get a job in the computing field]'].value_counts()
value_counts.plot(kind='bar', color='purple')
plt.title('Most Common Motivations to Seek Computing Degree')
plt.xlabel('Motivation Level')
plt.ylabel('Count')
plt.xticks(rotation=0)   
plt.show()
```

```python
value_counts.plot(kind='barh', color='skyblue')
plt.title('Frequency of Majors')
plt.xlabel('Count')
plt.ylabel('Major')
plt.show()
```
