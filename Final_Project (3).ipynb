### In which year did the United States have the highest per capita wind energy consumption and the highest share of solar energy in electricity generation?

#### Importing plots in pandas

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))  # Optional: Adjust the size of the plot
plt.plot(Energy_filtered['year'], Energy_filtered['wind_energy_per_capita'], marker='o', color='blue', linestyle='-', label='Wind Energy per Capita')

plt.title('Wind Energy per Capita in the USA (Year-wise)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Wind Energy per Capita (Units)', fontsize=12)

 
plt.grid(True)

 
plt.show()
```
